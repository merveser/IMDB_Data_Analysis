  <h1 style="color: #FFFFFF;  margin: 0;">VARIABLE DESCRIPTION</h1>


- **Series_Title**: The title of the movie or TV series.
- **Released_Year**: The year the movie or TV series was released.
- **Runtime**: The total runtime of the movie or TV series, usually given in minutes.
- **Genre**: The genre or categories the movie or TV series falls into, such as Action, Comedy, Drama, etc.
- **IMDB_Rating**: The rating of the movie or TV series on IMDb, typically on a scale from 1 to 10.
- **Meta_score**: The Metascore from Metacritic, which aggregates reviews and gives a score from 0 to 100.
- **Director**: The director of the movie or TV series.
- **No_of_Votes**: The number of votes the movie or TV series has received on

**Categorical Variables**: Series_Title, Genre, Director

**Numerical Variable**: Released_Year, Runtime, IMDB_Rating, Meta_score, No_of_Votes

<div style="background-color: #87CEEB; border: 1px solid #87CEEB; padding: 10px; border-radius: 15px; text-align: center;">
  <h1 style="color: #FFFFFF;  margin: 0;">HISTOGRAM</h1>

```python
def plot_hist(variable):
    plt.figure(figsize=(9, 5))
    plt.bar(df_first_10['Series_Title'], df_first_10[variable].dropna(), color='skyblue', edgecolor='k')
    plt.xlabel('Series Title')
    plt.ylabel(variable)
    plt.title(f"Histogram of {variable}")
    plt.xticks(rotation=45, ha='right')
    plt.show()
```

```python
numericVar = ["Released_Year", "Runtime", "IMDB_Rating", "Meta_score", "No_of_Votes", "Gross"]
for n in numericVar:
    plot_hist(n)
```
